# IMPORTING IMPORTANT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# LOADING DATASET

```python
df = pd.read_csv("kaggle-dataset-7.csv")
```

# SPLITING VARIABLES AND PLOTING THE DATASET

```python
x = df["answers"]
y = df["followers"]
plt.scatter(x,y,label = "stars" , color = "m" , marker = "*" , s=30)
plt.xlabel('answers')
plt.ylabel('followers')
plt.show()
```

```python
dataset = pd.read_csv("kaggle-dataset-7.csv")
```

```python
X = dataset.iloc[:, [2]].values
y = dataset.iloc[:, [3]].values
```

```python
from sklearn.model_selection import train_test_split 
```

# SPLITTING DATA INTO TRAIN AND TEST DATA

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.9, random_state = 0)
```

# IMPORTING THE REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

```

# PLOTING THE DATASET

```python
plt.scatter(X_test, y_test, color = 'red', marker= "*")
plt.plot(X_test, regressor.predict(X_test), color = 'blue')
plt.title('Quora DATA Analysis')
plt.xlabel('Answers')
plt.ylabel('Followers')
plt.show()
```

# FINAL ANALYSIS

```python
count=0
for i in range (len(y_test)):
    if(abs(y_test[i]-y_pred[i])<10):
        count=count+1
print("Number of cases following the theory that more followers will result in more Answers to your quetion is: ")
print((count/900)*100,"%")
```
